### K-means

- cílem je minimalizovat rozptyl uvnitř shluků a maximalizovat rozptyl mezi shluky iterativním přiřazováním datových bodů ke shlukům a ktrualizací centroidů shluků
- poradí si s velkým množstvím dat

```python
import matplotlib.pyplot as plt

# Seznam počtu klastrů
k_values = list(range(1, 11))
inertias = []

for k in k_values:
    model = KMeans(n_clusters=k, n_init='auto')
    model.fit(df)
    inertias.append(model.inertia_)
# inertias – výsledky z KMeans

fig, ax = plt.subplots()
ax.plot(k_values, inertias, '-o')
ax.set_xlabel('Počet shluků [k]')
ax.set_ylabel('Setrvačnost (inertia)')
ax.set_xticks(k_values)
plt.title('Elbow metoda')
plt.grid(True)
plt.show()
```

### Redukce dimenzionality model K-means

```python
from sklearn.decomposition import PCA

features = df.select_dtypes(include='number').columns.tolist()
features.remove('Total score')

X = df[features]

# K-means s k=4
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
df["Cluster"] = kmeans.fit_predict(df_scaled)

# Redukce dimenzionality pomocí PCA pro vizualizaci
# PCA hledá nové osy v prostoru dat, které zachytí co nejvíce variability (rozptylu)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(df_scaled)
df['PCA1'] = X_pca[:, 0]
df['PCA2'] = X_pca[:, 1]

# Vykreslení výsledku
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='PCA1', y='PCA2', hue='Cluster', palette='Set2', s=100)
plt.title('K-means clustering (k=4)')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.legend(title='Cluster')
plt.grid(True)
plt.tight_layout()
plt.show()
```
